SEARCHING AN ELEMENT IN ARRAY

```python
from array import*
```

```python
array1=array('i',[1,2,3,4,5,6,7])

def searchelement(array , value):
    for i in array:
        if i ==value:
            return array.index(value)
    return "The element does not exist"
print(searchelement(array1 , 5))
```
