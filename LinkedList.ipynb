```python
from random import randint
#CREATING NODE CLASS
class Node:
    def __init__(self, value= None):
        self.value = value
        self.next = None
        self.prev = None
    def __str__ (self):#CREATING NODE CLASS PRINTABLE
        return str(self.value)
#CREATED LL
class Linkedlist:
    def __init__(self,values = None):
        self.head = None
        self.tail = None
    def __iter__(self):
        curnode = self.head
        while curnode:
            yield curnode
            curnode = curnode.next
    def __str__(self):
        values = [str(x.value) for x in self]
        return '->'.join(values)
    def __len__(self ):
        result = 0
        node = self.head
        while node:
            result +=1
            node = node.next
        return result
    def add(self , value):
        if self.head is None:
            newnode = Node(value)
            self.head = newnode
            self.tail = newnode
        else:
            self.tail.next = Node(value)
            self.tail = self.tail.next
        return self.tail
# TO GENERATE A LL BY PROVIDING THE MINIMUM AND MAXIMUM VALUES OF THE LINKED LIST BY USING RANDOM FUNCTION.
    def generator (self, n , min_v , max_v):
        self.head = None
        self.tail = None
        for i in range(n):
            self.add(randint(min_v , max_v))
        return self
        
customll = Linkedlist()
customll.generator(10,0,99)
print(customll)
print(len(customll))
```
