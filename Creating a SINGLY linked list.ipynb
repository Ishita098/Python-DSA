```python
#CREATING A SINGLY LINKED LIST
```

```python
class Node:
    def __init__(self , value =None):
        self.value = value
        self.next = None           #O(1)

class Slinkedlist:
    def __init__(self):
        self.head =None            #O(1)
        self.tail =None             #O(1)
        
singlylinkedlist = Slinkedlist()
node1 = Node(1)
node2 = Node(2)

singlylinkedlist.head=node1
singlylinkedlist.head.next=node2     #O(1)
singlylinkedlist.tail=node2
```
