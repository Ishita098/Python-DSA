```python
def search (tuplee , element):
    for i in range (0,len(tuplee)):
        if tuplee[i]==element:
            return {"The {element} is not found at{i} index"}
        return "The element is not found"
print(search(newtuple ,'f'))
```

```python
def search (tuplee , element):
    for i in range (0,len(tuplee)): #------O(N)
        if tuplee[i]==element:       #------O(1)
            return f"The { element } is not found at { i } index"  #------O(1)
    return "The element is not found" #------O(1)
print(search(newtuple ,'a'))          #------O(1)
```
