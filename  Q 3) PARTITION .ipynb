```python
# Q 3) PARTITION - Write code to partition a linked list arounf a value x,such that all nodes less thean x come before all nodes greater than or equal to x .
```

```python
from LinkedList import LinkedList

def partition(ll,x):
    curnode = ll.head
    ll.tail = ll.head
    
    while curnode:
        nextnode = curnode.next
        curnode.next = None
        if curnode.value <x:
            curnode.next = ll.head
            ll.head = curnode
        else:
            ll.tail.next = curnode
            ll.tail = curnode
        curnode = nextnode
        
    if ll.tail.next is not None:
        ll.tail.next = None
customLL= LinkedList()
customLL.generate(10,0,99)
print(customLL)
partition(customLL , 30)
print(customLL)
```
