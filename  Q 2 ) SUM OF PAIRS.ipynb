# SUM OF PAIRS 

```python
def pairsum (nums , target):
    for i in range(len(nums)):
        for j in range (i+1 , len(nums)):
            if nums[i]==nums[j]:
                continue
            elif nums[i]+nums[j]==target:
                print ( i , j )
```

```python
mylist = [1,2,3,4,5,6,7,8]
```

```python
pairsum(mylist ,7 )
```
