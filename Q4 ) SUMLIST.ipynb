```python
from LinkedList import LinkedList

def sumlist (llA , llB):
    n1 = llA.head
    n2 = llB.head
    carry = 0
    ll = LinkedList()
    
    while n1 or n2:
        result = carry
        if n1:
            result +=n1.value
            n1 = n1.next
        if n2:
            result +=n2.value
            n2 = n2.next
        ll.add(int(result % 10))
        carry = result / 10
    return ll
llA = LinkedList()
llA.add(3)
llA.add(4)
llA.add(5)

llB = LinkedList()
llB.add(2)
llB.add(2)
llB.add(1)

print(llA)
print(llB)
print(sumlist(llA , llB))
```
