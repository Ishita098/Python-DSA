# Recursive vs iterative solutions

```python
def power2(n):
    if n==0:
        return 1
    else:
        power = power2(n-1)
        return power*2
```

```python
power2(3)
```

```python
def power3(n):
    if n==0:
        return 1
    else:
        power2 = power3(n-1)
        return power2*2
```
