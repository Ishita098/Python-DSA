```python
#2D ARRAY / INSERTING AN ELEMENT
```

```python
import numpy as np
```

```python
twodarray = np.array([[11,14,3,4],[3,4,5,1],[2,5,65,2],[6,8,9,9]])
print(twodarray)
```

```python
newtwo=np.insert(twodarray ,0,[[1,2,3,4]] ,axis=0)
print(newtwo)
```

```python
newtwo=np.insert(twodarray ,1,[[1,2,3,4]] ,axis=0)
print(newtwo)
```

```python
newtwo=np.insert(twodarray ,0,[[1,2,3,4]] ,axis=1)
print(newtwo)
```

```python
newtwo=np.insert(twodarray ,2,[[14,2,3,4]] ,axis=1)
print(newtwo)
```

```python
#2ND WAY (Using append we can add the new elements at the start or at the end of the row or column )
```

```python
newtwo=np.append(twodarray ,[[1,2,3,4]] ,axis=0)
print(newtwo)
```
