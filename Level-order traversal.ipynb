```python
import QueueLinkedList as queue
class Treenode:
    def __init__(self , data):
        self.data = data
        self.leftchild = None
        self.rightchild = None
    
newbt = Treenode("subjects")

leftchild = Treenode("maths")
physics = Treenode("physics")
maths = Treenode("maths")
leftchild.leftchild = physics
leftchild.rightchild = maths

rightchild = Treenode("science")
chem = Treenode("chem")
bio= Treenode("bio")
rightchild.leftchild = chem
rightchild.rightchild = bio
newbt.leftchild = leftchild
newbt.rightchild = rightchild

def levelordertraverdal(rootnode):
    if not rootnode:
        return
    else:
        customqueue = queue.Queue()
        customqueue.enqueue(rootnode)
        while not (customqueue.dequeue()):
            root = customqueue.dequeue()
            print(root.value.data)
            if (root.value.leftchild is not None):
                customqueue.enqueue(root.value.leftchild)
            if (root.value.rightchild is not None):
                customqueue.enqueue(root.value.rightchild)
                
levelordertraverdal(newbt)             

```
