```python
#DELETING ELEMENT FROM 2D ARRAY
```

```python
import numpy as np
twodarray = np.array([[1,2,3,4],[3,4,5,6],[4,6,7,2],[9,8,6,4]])
print(twodarray)
newarray =np.delete(twodarray ,1,axis=1)
print(newarray)
```

```python
twodarray = np.array([[1,2,3,4],[3,4,5,6],[4,6,7,2],[9,8,6,4]])
print(twodarray)
newarray =np.delete(twodarray ,0,axis=1)
print(newarray)
```

```python
twodarray = np.array([[1,2,3,4],[3,4,5,6],[4,6,7,2],[9,8,6,4]])
print(twodarray)
newarray =np.delete(twodarray ,1,axis=0)
print(newarray)
```

```python
twodarray = np.array([[1,2,3,4],[3,4,5,6],[4,6,7,2],[9,8,6,4]])
print(twodarray)
newarray =np.delete(twodarray ,3,axis=0)
print(newarray)
```
