```python
listz = [10,20,30,40,50,60,70,80,90]
```

```python
#linear search

def searchinlist(list ,value):
    for i in list:
        if i==value:
            return listz.index(value)
    else:
        return " The value does not exist"
```

```python
print (searchinlist(listz ,50))
```
