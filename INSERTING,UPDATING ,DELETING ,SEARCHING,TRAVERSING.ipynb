```python
#TRAVERSE
```

```python
mydict ={'name': 'edy', 'age': 28, 'address': 'london'}
```

```python
def traverse(dict):
    for key in dict:
        print(key,dict[key])
traverse(mydict)
```

```python
#SEARCHING IN DICT
```

```python
mydict ={'name': 'edy', 'age': 28, 'address': 'london'}
```

```python
def searchdict (dict ,value ):
    for key in dict:
        if dict[key]==value:
            return key,value
    return "The value does not exist"
print(searchdict(mydict,"london"))
```
